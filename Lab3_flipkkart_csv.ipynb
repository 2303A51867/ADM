```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the dataset
url = '/content/flipkart_com-ecommerce_sample.csv'
data = pd.read_csv(url)

# Display the first few rows of the dataset
print(data.head())

# Generate statistical summary of the dataset
summary = data.describe(include='all')
print(summary)
```

```python

# Plot histogram for 'retail_price' column
plt.hist(data['retail_price'].dropna(), bins=20, edgecolor='black', color='skyblue')
plt.title('Retail Price Distribution')
plt.xlabel('Retail Price')
plt.ylabel('Frequency')
plt.show()
```

```python
# Bar chart for 'brand' (Top 10 brands by count)
plt.figure(figsize=(10, 6))
data['brand'].value_counts().head(10).plot(kind='bar', color='lightcoral')
plt.title('Top 10 Brands by Count')
plt.xlabel('Brand')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.show()
```

```python
# Box plot for 'retail_price' to visualize outliers and distribution
plt.figure(figsize=(6, 4))
plt.boxplot(data['retail_price'].dropna(), patch_artist=True, boxprops=dict(facecolor='lightgreen'))
plt.title('Box Plot of Retail Price')
plt.ylabel('Retail Price')
plt.show()
```

```python
# Pairplot to visualize relationships between numerical features
sns.pairplot(data[['retail_price', 'discounted_price', 'product_rating']].dropna())
plt.show()
```

```python
data['product_rating'] = pd.to_numeric(data['product_rating'], errors='coerce')
correlation_matrix = data[['retail_price', 'discounted_price', 'product_rating']].corr()
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', cbar=True)
plt.title('Correlation Heatmap')
plt.show()
```

```python
sns.countplot(x='is_FK_Advantage_product', data=data, palette='Set2')
plt.title('FK Advantage Product Count')
plt.xlabel('FK Advantage Product')
plt.ylabel('Count')
plt.show()
```

```python
# Violin plot for 'retail_price' across 'brand' (Top 5 brands)
top_brands = data['brand'].value_counts().head(5).index
filtered_data = data[data['brand'].isin(top_brands)]
sns.violinplot(x='brand', y='retail_price', data=filtered_data, palette='muted')
plt.title('Retail Price Distribution by Top 5 Brands')
plt.xticks(rotation=45)
plt.show()
```
