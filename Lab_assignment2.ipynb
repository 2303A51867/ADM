```python
#1 histogram
import pandas as pd
import matplotlib.pyplot as plt
data=pd.read_csv('/content/diabetes.csv')
plt.hist(data['Age'], bins=40, color='skyblue',edgecolor='black')
plt.title('Age Distribution')
plt.xlabel('Age')
plt.ylabel('Frequeny')
plt.show()
```

```python
#2 statistical summary of dataset
import seaborn as sns
plt.figure(figsize=(12, 5))
plt.subplot(1, 2, 1)
sns.scatterplot(x=data["Glucose"], y=data["BMI"], hue=data["Outcome"], palette="coolwarm")
plt.title("Glucose vs. BMI")
plt.subplot(1, 2, 2)
sns.scatterplot(x=data["Age"], y=data["Insulin"], hue=data["Outcome"], palette="viridis")
plt.title("Age vs. Insulin")
plt.tight_layout()
plt.show()
```

```python
#3 Heatmap
plt.figure(figsize=(10, 8))
sns.heatmap(data.corr(), annot=True, cmap="coolwarm", square=True)
plt.title("Correlation Heatmap")
plt.show()
```
